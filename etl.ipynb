```python
import pandas as pd
```

```python
df = pd.read_csv('./Car_sales.csv')
df.head()
```

```python
df.info()
```

```python
null_counts = df.isnull().sum()
print(null_counts)
```

```python
df = df.drop(columns=['Power_perf_factor', 'Latest_Launch'])
```

```python
df['__year_resale_value'] = df['__year_resale_value'].fillna(df['__year_resale_value'].mean()).round(3)
df['Price_in_thousands'] = df['Price_in_thousands'].fillna(df['Price_in_thousands'].mean()).round(3)
df['Engine_size'] = df['Engine_size'].fillna(df['Engine_size'].mean()).round(2)
df['Horsepower'] = df['Horsepower'].fillna(df['Horsepower'].mean()).astype(int)
df['Wheelbase'] = df['Wheelbase'].fillna(df['Wheelbase'].mean()).round(1)
df['Width'] = df['Width'].fillna(df['Width'].mean()).round(1)
df['Length'] = df['Length'].fillna(df['Length'].mean()).round(1)
df['Curb_weight'] = df['Curb_weight'].fillna(df['Curb_weight'].mean()).round(3)
df['Fuel_capacity'] = df['Fuel_capacity'].fillna(df['Fuel_capacity'].mean()).round(1)
df['Fuel_efficiency'] = df['Fuel_efficiency'].fillna(df['Fuel_efficiency'].mean()).round(1)
null_counts = df.isnull().sum()
print(null_counts)
```

```python
df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
df.columns = df.columns.str.lower()
df = df.rename(columns={'__year_resale_value': 'year_resale_value'})
df = df.rename(columns={'wheelbase': 'wheel_base'})
df.head()
```

```python
df.to_csv('./clean_car_sales.csv', index=False)
```
